지금은 AI 시스템을 활용한 자동화가 세상을 놀라게 하고 있다. 많이 고민하며 한줄 한줄 코딩했던 프로그래밍조차도 최근에는 바이브 코딩을 통하여 프로그램 언어를 모르더라도 아이디어만으로도 프로그래밍을 AI가 작성해 주는 실정이다. 하지만, 프로그래밍에 대한 가장 기본적인 것을 직접 할 수 있는 상태에서 바이브 코딩을 바라보고 활용하면 그렇지 않을 때와는 다를 것이다. 또한 프로그래밍을 잘 살펴보면 논리로 구성되어 있다. 즉, 수학이 가장 기본입니다.

강의의 초점은 수학공학도구인 Geogebra를 활용하여 프로그래밍의 기본을 익히고 산출물을 만드는 방법을 제시합니다.

프로그래밍의 기본은 조건과 반복으로 구성되어 있다. 이 기본적인 기능을 익히고 조금 복잡한 개념이겠지만, 이런 논리적인 구성을 가진 객체들 상호 관계를 구성하면 수학적 아이디어를, 프로그래밍을 통하여 더 많은 표현을 할 수 있다.

또한, 기록해야 다음 단계의 표현이 가능한데, 프로그래밍에서 저장하는 방식을 이해할 필요가 있다. 저장하는 대상은 숫자, 문자 등 간단한 것부터, 이런 것들이 모인 복잡한 대상들이다. 또한 저장하는 순서의 유무에 따라 집합 또는 순서쌍 방식으로 저장법을 익혀야 합니다.

여기에 Geogebra에서 제공하는 수학적인 점, 선분, 원, 타원 등 수학적인 객체와 이와 같은 기본적인 것을 결합하면 수학적 아이디어의 프로그래밍을 통하여 표현할 수 있다.

그동안, 수업의 필요로 수학 수업 자료를 많이 만들었다. 단편적인 기능설명보다는 실제 수업 자료를 통해서 위와 같은 기능을 익히겠다.

실제 자료에 다음과 같은 명령어를 익힐 수 있습니다. 중복되는 것은 제외하고 명령어 설명을 하고 자료를 제시합니다. 처음에는 이런 것을 어디에 쓸까? 라는 생각도 하겠지만, 예시 자료에서 쓰이는 것을 익히면 그 쓰임과 응용을 같이 익힐 수 있을 것입니다.

* ```
  Unique[<list>]
  <list>의 중복 원소를 제거한 <list>, <list>를 set처럼 중복 없게 만들었지만, 속성은 여전히 <list>이다.

* ```
  Sequence[f(t),t,m,n]
  t를 m부터 n까지 순차적으로 증가시키며 함수 f(t)의 결과를 순차적으로 원소로 구성한 list를 만든다.
  ```


* ```
  Element[<list>,t]
  <list>의 t+1번째 원소를 나타낸다.
  ```


* ```
  위의 명령어를 합성한 형태 예시
  Sequence[CountIf[x==Element[<listA>,t],<listB>],t,m,n]
    <listA>에 m번째에서 n번째 각각의 원소가 <listB>에 있는 개수를 순차적으로 <list>로 만들라는 뜻이다.

  합성된 명령어를 아래와 같이 풀어서 이해하도록 하자.
  Sequence[
            CountIf[
                    x==Element[<listA>,t],
                    <listB>
                   ],
            t,m,n]
  ```


하지만, Geogebra 스크립트 방식의 프로그래밍을 줄 바꿈 하면서 가독성 있게 나타낼 수는 없으므로 이해를 위해서만 위와 같이 풀어서 쓴 것임을 주의하기를 바란다.

삼각형의 페르마 점(Fermat point)은 삼각형의 세 꼭짓점에서 거리의 합이 최소가 되는 점을 말한다. 이를 두 가지 시각에서 접근하려고 한다. 하나의 시각은 작도를 활용하여 페르마 점을 작도하는 방법이다. 다른 시각은 인공지능에 쓰이는 경사하강법을 활용하여 페르마 점에 다가가는 방법이다. 이 두 가지를 수학공학도구인 Geogebra를 활용하여 만드는 자료입니다.
